```python
%matplotlib inline

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
folder='../datasets'
filename='bf_2019-07-14 07:01:13.030863.csv'
```

```python
import csv

time_series = []
with open(f"{folder}/{filename}", 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        for x in row:
            try:
                time_series.append(float(x))
            except ValueError:
                pass
            
print(time_series[:10], time_series[-10:])
```

```python
from math import exp
def linear(x):
    return x
def exponent(x):
    return exp(x/2)
def power_of_1_5(x):
    return x ** 1.5
def power_of_2(x):
    return x ** 2
def power_of_3(x):
    return x ** 3
def power_of_4(x):
    return x ** 4
def power_of_10(x):
    return 10 ** x
def two_to_the_power_of(x):
    return 2 ** x

funcs = [linear, exponent, power_of_1_5, power_of_2, power_of_3, power_of_4, power_of_10, two_to_the_power_of]
```

```python
for func in funcs:
    plt.plot([x for x in time_series[:10]], label=f"{filename} first 10")
    plt.plot([func(x) for x in range(10)], label=func.__name__)
    plt.legend()
    plt.show()
```

```python
for func in funcs:
    plt.plot([x for x in time_series[:100]], label=f"{filename} first 100")
    plt.plot([func(x) for x in range(100)], label=func.__name__)
    plt.legend()
    plt.show()
```

```python
for func in funcs:
    plt.plot([x for x in time_series[1:10]], label=f"{filename} first 10")
    plt.plot([func(x) for x in range(1,10)], label=func.__name__)
    plt.xscale("log")
    plt.yscale("log")
    plt.legend()
    plt.show()
```

```python
for func in funcs:
    plt.plot([x for x in time_series[1:100]], label=f"{filename} first 100")
    plt.plot([func(x) for x in range(1,100)], label=func.__name__)
    plt.xscale("log")
    plt.yscale("log")
    plt.legend()
    plt.show()
```

```python
rng = 10
from math import sqrt, atan2
r = [sqrt(x1 ** 2 + time_series[x1] ** 2 ) for x1 in range(rng)]
teta = [atan2(time_series[x1], x1) for x1 in range(rng)]
plt.polar(teta, r)
plt.title(f"polar projection - {filename} range < {rng}")
plt.show()
```

```python
from math import sqrt, atan2
r = [sqrt(x1 ** 2 + time_series[x1] ** 2 ) for x1 in range(len(time_series))]
teta = [atan2(time_series[x1], x1) for x1 in range(len(time_series))]
plt.polar(teta, r)
plt.title(f"polar projection - {filename}")
plt.show()
```
