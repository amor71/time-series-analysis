```python
%matplotlib inline

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
def f(x):
    return x ** 2
def g(x):
    return x ** 3
def h(x):
    return x ** 4
def p(x):
    return 10 ** x
def t(x):
    return 2 ** x
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 100, 1000)
ax.plot(x, f(x))
plt.title("linear scale - power of 2")
plt.show()
```

```python
ax = plt.axes()
x = np.linspace(0, 100, 1000)
ax.plot(x, f(x))
ax.plot(x, g(x))
plt.title("linear scale - power of 2 & 3")
plt.show()
```

```python
ax = plt.axes()
x = np.linspace(0, 100, 1000)
ax.plot(x, f(x))
ax.plot(x, g(x))
ax.plot(x, h(x))
plt.title("linear scale - power of 2 & 3 & 4")
plt.show()
```

```python
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, f(x))
ax.plot(x, g(x))
ax.plot(x, h(x))
ax.plot(x, t(x))
plt.title("linear scale - power of 2 & 3 & 4 and 2 to the power of x")
plt.show()
```

```python
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, f(x), label="x^2")
ax.plot(x, g(x), label="x^3")
ax.plot(x, h(x), label="x^4")
ax.plot(x, p(x), label="10^x")
ax.plot(x, t(x), label="2^x")
ax.legend()
plt.title("linear scale - power of 2 & 3 & 4 and 2 & 10 to power of x")
plt.show()
```

```python
ax = plt.axes()
x = np.linspace(1, 10, 1000)
ax.plot(x, f(x), label="x^2")
ax.plot(x, g(x), label="x^3")
ax.plot(x, h(x), label="x^4")
ax.plot(x, p(x), label="10^x")
ax.plot(x, t(x), label="2^x")
ax.legend()
ax.set_yscale("log")
plt.title("log scale - power of 2 & 3 & 4 and 2 & 10 to power of x")
plt.show()
```

```python
ax = plt.axes()
x = np.linspace(1, 100, 1000)
ax.plot(x, f(x), label="x^2")
ax.plot(x, g(x), label="x^3")
ax.plot(x, h(x), label="x^4")
ax.legend()
ax.set_yscale("log")
ax.set_xscale("log")
plt.title("log log scale - power of 2 & 3 & 4 ")
plt.show()
```

```python
from math import sqrt, atan2
x = np.linspace(0, 10, 10000)
r = [sqrt(x1 ** 2 + f(x1) ** 2 ) for x1 in x]
teta = [atan2(f(x1), x1) for x1 in x]
plt.polar(teta, r)
plt.title("polar projection - power of 2")
plt.show()
```

```python
from math import sqrt, atan2
x = np.linspace(0, 2, 1000)
r0 = [x1 for x1 in x]
r1 = [sqrt(x1 ** 2 + f(x1) ** 2 ) for x1 in x]
rd = [x1 - sqrt(x1 ** 2 + f(x1) ** 2 ) for x1 in x]
teta0 = [atan2(x1, x1) for x1 in x]
teta1 = [atan2(f(x1), x1) for x1 in x]
plt.polar(teta0, r0, label='x')
plt.polar(teta1, r1, label='x^2')
plt.title("polar projection - power of 1, 2")
plt.show()
```

```python
from math import sqrt, atan2
x = np.linspace(0, 2, 1000)
r0 = [x1 for x1 in x]
r1 = [sqrt(x1 ** 2 + f(x1) ** 2 ) for x1 in x]
r2 = [sqrt(x1 ** 2 + g(x1) ** 2 ) for x1 in x]
teta0 = [atan2(x1, x1) for x1 in x]
teta1 = [atan2(f(x1), x1) for x1 in x]
teta2 = [atan2(g(x1), x1) for x1 in x]
plt.polar(teta0, r0, label='x')
plt.polar(teta1, r1, label='x^2')
plt.polar(teta2, r2, label='x^3')
plt.title("polar projection - power of 1, 2, 3")
plt.show()
```
