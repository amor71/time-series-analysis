```python
%matplotlib inline

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
folder='../datasets'
filename='wn_2019-07-28 20:51:38.907133.csv'
```

```python
import csv

time_series = []
with open(f"{folder}/{filename}", 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        for x in row:
            try:
                time_series.append(float(x))
            except ValueError:
                pass
            
print(time_series[:10], time_series[-10:])
```

```python
plt.plot(time_series)
plt.title(f"{filename} time series")
plt.show()
```

```python
deltas = [time_series[t] - time_series[t - 1] for t in range(1, 100)]
plt.plot(deltas)
plt.title(f"Delta time series")
plt.show()
```

```python
from math import ceil
plt.title("Values distribution")
count, bins, ignored  = plt.hist(time_series, int(ceil(max(time_series))-ceil(min(time_series))), color='green')
```

```python
from math import ceil
plt.title("Deltas distribution")
count, bins, ignored  = plt.hist(deltas, int(ceil(max(deltas))-ceil(min(deltas))), color='green')
```
