```python
%matplotlib inline

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
folder='../datasets'
filename='gold-prices.csv'
```

```python
import csv
import locale
locale.setlocale( locale.LC_ALL, 'en_US.UTF-8' )

time_series = []
with open(f"{folder}/{filename}", 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        for x in row:
            try:
                time_series.append(locale.atof((x)))
            except ValueError:
                print(x)
                pass
            
print(time_series[:10], time_series[-10:])
```

```python
time_series = time_series
```

```python
plt.plot(time_series)
plt.title(f"{filename} time series")
plt.show()
```

```python
print(max(time_series))
```

```python
deltas = [time_series[t] - time_series[t - 1] for t in range(1, 100)]
plt.plot(deltas)
plt.title(f"Prefix Delta time series")
plt.show()
```

```python
deltas = [time_series[t] - time_series[t - 1] for t in range(1, len(time_series))]
plt.plot(deltas)
plt.title(f"Full Delta time series")
plt.show()
```

```python
from math import ceil
plt.title("Values distribution")
count, bins, ignored  = plt.hist(time_series, int(ceil(max(time_series))-ceil(min(time_series))), color='green')
```

```python
from math import ceil
plt.title("Deltas distribution")
count, bins, ignored  = plt.hist(deltas, int(ceil(max(deltas))-ceil(min(deltas))), color='green')
```

```python
from math import log
log_return = [log(1.0 * time_series[t + 1] / time_series[t]) for t in range(0, len(time_series)-1)]
plt.plot(log_return)
plt.title("logarithmic return")
plt.show()
```
